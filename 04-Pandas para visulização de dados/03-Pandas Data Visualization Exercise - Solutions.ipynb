Este é apenas um exercício rápido para você revisar os vários gráficos que mostramos anteriormente. Use **df3** para replicar os gráficos a seguir.

```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

**Recrie este scatter plot de `b` vs `a`. Observe a cor e o tamanho dos pontos. Observe também o tamanho da figura. Veja se você pode descobrir como esticá-lo de forma semelhante. Lembre-se de sua seção matplotlib...**

```python
df3.plot.scatter(x='a',y='b',c='red',s=50,figsize=(12,3))
```

**Crie um histograma da coluna 'a'.**

```python
df3['a'].plot.hist()
```

**Esses gráficos são bons, mas não parecem muito polidos. Use style sheets para definir o estilo como 'ggplot' e refaça o histograma de cima. Descubra também como adicionar mais compartimentos a ele.**

```python
plt.style.use('ggplot')
```

```python
df3['a'].plot.hist(alpha=0.5,bins=25)
```

**Crie um boxplot comparando as colunas `a` e `b`.**

```python
df3[['a','b']].plot.box()
```

**Crie um gráfico kde da coluna 'd'**

```python
df3['d'].plot.kde()
```

**Descubra como aumentar a largura da linha e tornar o estilo de linha (linestyle) tracejada. (Nota: Você normalmente não traça uma linha de plotagem do kde)**

```python
df3['d'].plot.density(lw=5,ls='--')
```

**Crie um gráfico de área de todas as colunas apenas para as linhas até 30.**

```python
#df3.ix[0:30].plot.area(alpha=0.4)
df3.iloc[0:30].plot.area(alpha=0.4)
```


## Desafio Bônus!

Observe que você pode achar isso muito difícil, consulte as soluções se não conseguir descobrir!
**Observe como a legenda em nossa figura anterior se sobrepôs a alguns diagramas reais. Você consegue descobrir como exibir a legenda fora do gráfico, conforme mostrado abaixo?**

**Tente pesquisar no Google por um bom link de stackoverflow sobre este tópico. Se você não conseguir encontrá-lo por conta própria - [use este como uma dica.](http://stackoverflow.com/questions/23556153/how-to-put-legend-outside-the-plot-with-pandas )**

```python
f = plt.figure()
df3.ix[0:30].plot.area(alpha=0.4,ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
