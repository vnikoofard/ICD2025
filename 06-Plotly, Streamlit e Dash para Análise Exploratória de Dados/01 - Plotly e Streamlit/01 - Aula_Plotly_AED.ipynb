Mapa de cor

*Um dicionário {'variável_1':'cor_1', 'variável_2':'cor_2'...} pode ser criado para inserir cores de sua preferência*

```python
mapa_cor_iris = {"setosa":"black",
                 "versicolor":"yellow",
                 "virginica":"white"}

fig = px.scatter(data_frame = iris,
                  x= "comprimento_petala",
                  y = "largura_petala",
                  color = "espécie",
                  color_discrete_map = mapa_cor_iris,
                  symbol="espécie",
                  size="comprimento_sepala")

fig.show()
```

## Conjunto de Dados: Olimpíadas

Importando os dados

```python
olimpiadas = px.data.medals_long()
olimpiadas
```

Renomeando as colunas

```python
olimpiadas = olimpiadas.rename(columns={"nation":"nação",
                           "medal":"medalha",
                           "count":"quantidade"})
mapeamento_nação = {
    'South Korea': 'Coréia do Sul'
    }
mapeamento_medalhas = {
    'gold': 'Ouro',
    'silver': 'Prata',
    'bronze': 'Bronze'
    }
olimpiadas['nação'] = olimpiadas['nação'].replace(mapeamento_nação)
olimpiadas['medalha'] = olimpiadas['medalha'].replace(mapeamento_medalhas)
olimpiadas
```

Explorando os dados

```python
olimpiadas.info()
```

Resumo estatístico

```python
olimpiadas.describe()
```

Resumo estatístico categórico

*Com df.describe(include="O"), obtemos um relatório estatístico de variáveis categóricas (objects)*

```python
olimpiadas.describe(include="O")
```

Plotando gráfico de barras

*Com px.bar(data_frame=df, x="coluna_x", y="coluna_y") criamos um gráfico de barras*

```python
fig = px.bar(
    data_frame = olimpiadas,
    x = "nação",
    y = "quantidade"
)

fig.show()
```

Aplicar cores

*Com o argumento color="coluna_cor" incluimos cor às colunas*

```python
fig = px.bar(data_frame = olimpiadas,
              x="nação",
              y="quantidade",
              color="medalha",
              barmode = "group")
fig.show()
```

Escolhendo nossas cores, agrupando e aplicando textos

*Com o argumento barmode, conseguimos organizar as colunas de forma empilhada (padrão) ou lateralizadas (group). Utilizando text_auto, conseguimos visualizar os valores de cada coluna*

*Além disso podemos incluir título e nomes ao eixo x e y*

```python
cores = {"Ouro":"gold",
         "Prata":"silver",
         "Bronze":"brown"}

fig = px.bar(olimpiadas,
              x="nação",
              y="quantidade",
              color="medalha",
              color_discrete_map=cores,
              barmode="group",
              text_auto=True)

fig.update_layout(title="Análise de Medalhas dos Países",
                  xaxis_title="Países",
                  yaxis_title="Quantidade de Medalhas",
                  legend_title="Medalha")

fig.show()
```

```python
acoes = px.data.stocks()
acoes.head()
```

Plot com 2 eixos y

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=acoes["date"], y=acoes["GOOG"], name="yaxis data"),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=acoes["date"], y=acoes["AAPL"], name="yaxis2 data"),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text="Double Y Axis Example"
)

# Set x-axis title
fig.update_xaxes(title_text="xaxis title")

# Set y-axes titles
fig.update_yaxes(title_text="<b>primary</b> yaxis title", secondary_y=False)
fig.update_yaxes(title_text="<b>secondary</b> yaxis title", secondary_y=True)

fig.show()
```

## Conjunto de Dados Numéricos e Categóricos: Restaurante

*Conjunto de dados de um restaurante que recebe semanalmente diversos clientes.*

```python
restaurante = px.data.tips()
restaurante.head()
```

```python
restaurante = restaurante.rename(columns={"total_bill":"conta_total",
                           "tip":"gorjeta",
                           "sex":"genero",
                            "smoker":"fumante",
                            "day":"dia",
                            "time":"horário",
                            "size":"tamanho"})
mapeamento_sexo = {
    'Male': 'Homem',
    'Female': 'Mulher'
    }

mapeamento_fumante = {
    'Yes': 'Sim',
    'No': 'Não'}

mapeamento_dia = {
    'Sun':'Dom',
    'Sat':'Sab',
    'Thur':'Qui',
    'Fri':'Sex'
}

mapeamento_ref = {
    'Dinner':'Jantar',
    'Lunch':'Almoço'
}

restaurante['genero'] = restaurante['genero'].replace(mapeamento_sexo)
restaurante['fumante'] = restaurante['fumante'].replace(mapeamento_fumante)
restaurante['dia'] = restaurante['dia'].replace(mapeamento_dia)
restaurante['horário'] = restaurante['horário'].replace(mapeamento_ref)
restaurante.head()
```

Explorando os dados

```python
restaurante.info()
```

Informações estatísticas numéricas

```python
restaurante.describe()
```

Informações estatísticas categóricas

```python
restaurante.describe(include="O")
```

**Qual foi o total arrecadado em contas pelo restaurante?**

*com o comando df["coluna"].sum() obtemos o somatório de todos os valores daquela coluna*

```python
total_contas = restaurante["conta_total"].apply("sum")
total_contas
```

**Qual o total arrecadado apenas com gorjetas pelo restaurante?**

```python
total_gorjetas = restaurante["gorjeta"].sum()
total_gorjetas
```

**Qual a razão entre gorjetas e contas?**

```python
gorjetas_contas = total_gorjetas/total_contas
gorjetas_contas
```

**Em quais dias da semana o restaurante abre?**

*o comando df['coluna'].unique() retorna todas as variáveis únicas de uma coluna*

```python
dias = restaurante["dia"].unique()
dias
```

**Quais são os dias mais e menos movimentados do restaurante?**

*com value_counts() podemos realizar a contagem de todas as variáveis de uma coluna. Além disso podemos organizar os valores com sort_values()*

```python
dias_movimentados = restaurante["dia"].value_counts().reset_index(name="Qtd")
dias_movimentados
```

**Quais são as refeições mais e menos comuns no restaurante?**

```python
refeições = restaurante["horário"].value_counts()
refeições
```

**Quantos homens e mulheres frequentam o restaurante?**

*com .reset_index(name="nome"), reorganizamos um data frame*

```python
homens_mulheres = restaurante["genero"].value_counts().reset_index(name="Qtd")
homens_mulheres
```

**Qual a razão entre homens e mulheres que frequentam o restaurante?**

```python
homens_mulheres_norm = restaurante["genero"].value_counts(normalize=True)
homens_mulheres_norm
```

**Quais são os gastos totais de homens e mulheres?**

*o comando df.groupby("coluna1")["coluna2"] agrupa todos os dados do data frame de acordo com valores únicos da coluna 1. A partir desse agrupamento, podemos entender a relação da coluna 1 com a coluna 2, através de somatórios, contagens e médias, por exemplo.*

```python
conta_genero = restaurante.groupby("genero")["conta_total"].sum().reset_index(name="Total")
conta_genero
```

```python
conta_genero["Gasto Pessoal"] = conta_genero["Total"] / homens_mulheres["Qtd"]
conta_genero
```

**Quantos fumantes e não fumantes frequentaram o restaurante?**

```python
fumantes = restaurante["fumante"].value_counts()
fumantes
```

**Quantos homens e mulheres fumam?**

*com df.groupby(["coluna1","coluna2",...]) podemos entender a correlação entre 2 ou mais colunas*

```python
genero_fuma = restaurante.groupby(["genero","fumante"]).size().reset_index(name="Qtd")
genero_fuma
```

Reunindo nossas análises em um formato gráfico

*podemos colorir cada coluna de acordo com o caso analisado*

*podemos colorir cada coluna de acordo com o caso analisado*

```python
mapa_cores = {"Sim":"green",
              "Não":"red" }

fig = px.bar(data_frame = genero_fuma,
              x="genero",
              y="Qtd",
              color="fumante",
              color_discrete_map=mapa_cores,
              barmode="group")

fig.update_layout(title="Homens e Mulheres - Fumantes",
                  xaxis_title="Gênero",
                  yaxis_title="Quantidade")

fig.show()
```

*com uma lista de variáveis ordenadas, junto aos argumentos em fig.update_xaxes(categoryorder="array", categoryarray=lista), podemos aplicar uma ordem personalizada das barras*

```python
mapa_ordem = ["Qui","Sex","Sab","Dom"]

mapa_cores = {"Qui":"gray",
              "Sex":"red",
              "Sab":"gray",
              "Dom":"gray"}

fig = px.bar(data_frame = dias_movimentados,
              x="index",
              y="Qtd",
              color = "index",
              color_discrete_map=mapa_cores,
              text_auto = True)

fig.update_xaxes(categoryorder="array",
                 categoryarray=mapa_ordem)

fig.update_layout(title="Dias Mais Movimentados",
                  xaxis_title="Dia",
                  yaxis_title="Quantidade")

fig.show()
```

```python
fig = px.bar(data_frame = mundo_populosos,
              x="país",
              y="pop",
              animation_frame="ano",
              animation_group="país",
              range_y = [0,1.35e9],
              color="país")

fig.update_xaxes(categoryorder="total descending")

fig.show()
```
