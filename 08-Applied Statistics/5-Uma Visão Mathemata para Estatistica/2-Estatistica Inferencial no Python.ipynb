# Estatística inferencial

A estatística inferencial é uma área da estatística que envolve o uso de amostras para fazer inferências ou generalizações sobre uma população maior. Ela permite que os estatísticos e pesquisadores tirem conclusões ou tomem decisões com base em informações limitadas, porém representativas, coletadas de uma amostra.

A estatística inferencial é amplamente utilizada em diversas áreas, como ciências sociais, negócios, medicina, economia, entre outras. Seu objetivo principal é estender os resultados obtidos em uma amostra para a população como um todo, levando em consideração a incerteza associada a essa generalização.

Para realizar inferências estatísticas, são utilizadas técnicas e métodos estatísticos, tais como:

1. Testes de hipóteses: São usados para tomar decisões sobre uma população com base em informações obtidas de uma amostra. Eles ajudam a determinar se uma afirmação ou suposição sobre a população é suportada pelos dados da amostra.

2. Intervalos de confiança: São intervalos estimados que fornecem uma faixa de valores prováveis para um parâmetro populacional. Esses intervalos são calculados a partir dos dados da amostra e fornecem uma medida de incerteza em torno da estimativa pontual.

3. Estimação pontual: Envolve a obtenção de uma única estimativa para um parâmetro populacional com base nos dados da amostra. Essa estimativa é usada para fornecer uma aproximação do valor real do parâmetro.

Essas são apenas algumas das técnicas utilizadas na estatística inferencial. O uso correto dessas técnicas permite que os pesquisadores tirem conclusões válidas sobre uma população com base em informações limitadas obtidas de uma amostra.

## O que é uma distribuição?

Uma distribuição é uma função que mostra todos os valores possíveis (ou eventos) de um dado fenômeno e a frequência (ou probabilidade) de cada evento. As distribuições podem ser representadas de várias maneiras, como tabelas, gráficos ou fórmulas matemáticas.

As distribuições podem ser divididas em duas categorias principais: distribuições de probabilidade e distribuições de frequência.

1. Distribuição de probabilidade: É uma função que mostra todas as probabilidades associadas aos possíveis valores de uma variável aleatória. As distribuições de probabilidade são usadas para descrever incertezas em experimentos aleatórios, como jogar um dado, lançar uma moeda, etc.

2. Distribuição de frequência: É uma função que mostra a frequência de ocorrência de cada valor de uma variável. As distribuições de frequência são usadas para resumir dados categóricos ou discretos.

### O que é uma distribuição de uniforme discreto?

A distribuição uniforme é uma distribuição de probabilidade discreto que é usada para modelar a probabilidade de ocorrência de eventos, onde todas das possibilidades são igualmente prováveis de ocorrer.

A distribuição uniforme é usada para modelar a probabilidade de ocorrência de eventos, onde todos os resultados possiveis são igualmente prováveis de ocorrer. Por exemplo, a distribuição uniforme pode ser usada para modelar a probabilidade de um dado cair em qualquer um dos seis lados.

### O que é uma distribuição de Bernoulli?

A distribuição de Bernoulli é uma distribuição de probabilidade discreta que expressa a probabilidade de um determinado evento ocorrer (sucesso ou fracasso) em um experimento de Bernoulli. Um experimento de Bernoulli é um experimento que possui as seguintes propriedades:

1. O experimento consiste em uma única tentativa.

2. A tentativa tem apenas dois resultados possíveis: sucesso ou fracasso.

3. A probabilidade de sucesso $(p)$ é a mesma para todas as tentativas.

A distribuição de Bernoulli é usada para modelar a probabilidade de ocorrência de um determinado evento em um experimento de Bernoulli. Por exemplo, a distribuição de Bernoulli pode ser usada para modelar a probabilidade de obter um sucesso em um único lançamento de moeda.


### O que é uma distribuição binomial?

A distribuição binomial é uma distribuição de probabilidade discreta que expressa a probabilidade de um determinado número de sucessos ocorrer em um determinado número de tentativas em um experimento binomial. Um experimento binomial é um experimento que possui as seguintes propriedades:

1. O experimento consiste em $n$ tentativas.

2. Cada tentativa tem apenas dois resultados possíveis: sucesso ou fracasso.

3. A probabilidade de sucesso ($p$) é a mesma para todas as tentativas.

A distribuição binomial é usada para modelar a probabilidade de ocorrência de um determinado número de sucessos em um experimento binomial. Por exemplo, a distribuição binomial pode ser usada para modelar a probabilidade de obter $5$ caras em $10$ lançamentos de moedas.

````python
# Parâmetros da distribuição binomial
n = 50  # Número de ensaios
p = 0.5  # Probabilidade de sucesso em cada ensaio

# Gerar valores para o eixo x
x = np.arange(0, n+1)

# Calcular os valores da distribuição binomial
y = np.random.binomial(n, p, size=n+1)

# Plotar a distribuição binomial
plt.figure(figsize=(5, 4))
plt.bar(x, y)

# Configurar o título e os rótulos dos eixos
plt.title('Distribuição Binomial')
plt.xlabel('Número de Sucessos')
plt.ylabel('Frequência')

# Exibir o gráfico
plt.show()
````


### O que é uma distribuição normal?

A distribuição normal é uma distribuição de probabilidade contínua que é simétrica em torno da média, mostrando que os dados próximos à média são mais frequentes em ocorrência do que os dados distantes da média. Além disso, a distribuição normal segue a curva de sino, o que significa que a maioria dos dados em uma distribuição normal está perto da média e os dados restantes estão próximos da média.

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} \cdot \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)
$$

A distribuição normal é uma das distribuições mais importantes na estatística. Ela é frequentemente usada em estatística descritiva para descrever e resumir dados. Além disso, muitas técnicas estatísticas assumem que os dados seguem uma distribuição normal, como o teste t, a análise de variância (ANOVA) e a regressão linear.

````python
# Parâmetros da distribuição normal
mu = 0  # Média
sigma = 1  # Desvio padrão

# Gerar valores para o eixo x
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

# Calcular os valores da distribuição normal
y = (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-(x-mu)**2/(2*sigma**2))

# Plotar a distribuição normal
plt.figure(figsize=(5, 4))
plt.plot(x, y)

# Configurar o título e os rótulos dos eixos
plt.title('Distribuição Normal')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')

# Exibir o gráfico
plt.show()
````

### Padronização

A padronização é o processo de transformar dados em uma distribuição normal padrão, que tem uma média de 0 e um desvio padrão de 1. A padronização é útil porque transforma os dados em uma distribuição normal padrão, independentemente da distribuição original dos dados. Isso permite que você compare facilmente valores de diferentes distribuições.

$$
\mathcal N(\mu, \sigma^2) \rightarrow \mathcal N(1,0)
$$

Essa transformação é possível via

$$
Z = \frac{x - \mu}\sigma \rightarrow Z \sim \mathcal N(1,0)
$$

### O que é uma distribuição de t de Student?

A distribuição $t$ é semelhante à distribuição normal, porém é caracterizada por ter caudas mais pesadas e ser mais achatada. Isso significa que a distribuição $t$ permite que valores extremos sejam mais prováveis do que na distribuição normal, especialmente quando o tamanho da amostra é pequeno.

Certamente! Aqui está a fórmula da distribuição t de Student em LaTeX:


A distribuição $t$ é usada principalmente quando o tamanho da amostra é pequeno e a variância populacional é desconhecida. É comumente aplicada em testes de hipóteses e na construção de intervalos de confiança para a média populacional, quando as condições de aplicação da distribuição normal não são atendidas.

A forma da distribuição $t$ é controlada pelos graus de liberdade ($df$ - degrees of freedom). Os graus de liberdade representam o número de observações independentes utilizadas para estimar uma quantidade desconhecida na distribuição. Quanto maior o número de graus de liberdade, mais a distribuição t se aproxima da distribuição normal.


A distribuição $t$ possui uma *função de densidade de probabilidade (PDF)* que descreve a probabilidade de obter um determinado valor. A média da distribuição t é zero e a variância é igual a df/(df-2) para df > 2.


$$
f(x) = \frac{{\Gamma\left(\frac{{\nu+1}}{2}\right)}}{{\sqrt{{\nu\pi}} \cdot \Gamma\left(\frac{\nu}{2}\right)}} \cdot \left(1+\frac{x^2}{\nu}\right)^{-\frac{{\nu+1}}{2}}
$$

Nessa fórmula:
- $f(x)$ representa a função de densidade de probabilidade da distribuição t de Student para um determinado valor $x$.
- $\nu$ é o número de graus de liberdade da distribuição t.
- $\Gamma$ é a função gama, que descreve uma generalização do fatorial para números reais positivos. A função gama é definida como $\Gamma(x) = (x-1)!$ para números inteiros positivos $x$. Para números reais positivos, a função gama é definida como $\Gamma(x) = \int_0^\infty t^{x-1}e^{-t}dt$.


A distribuição t é amplamente utilizada em várias áreas da estatística e fornece uma alternativa robusta quando as suposições da distribuição normal não são atendidas, especialmente em casos de amostras pequenas.


````python
from scipy.stats import t, norm

# Definindo os graus de liberdade para a distribuição t
df_values = [1, 3, 10, 30]

# Definindo o intervalo de valores para x
x = np.linspace(-4, 4, 1000)

# Plot da distribuição t e distribuição normal
plt.figure(figsize=(7, 5))
for df in df_values:
    plt.plot(x, t.pdf(x, df), label=f't (df={df})')


plt.plot(x, norm.pdf(x), label='Normal')

plt.title('Comparação da Distribuição t de Student e Normal')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.show()
````

### O que é uma distribuição de Poisson?

A distribuição de Poisson é uma distribuição de probabilidade discreta que modela a ocorrência de eventos raros e independentes em um intervalo fixo de tempo ou espaço. Ela é amplamente utilizada em áreas como estatística, teoria das filas, telecomunicações, entre outras.

A distribuição de Poisson é caracterizada por um único parâmetro, denotado por \(\lambda\), que representa a taxa média de ocorrência dos eventos raros. Os eventos devem ser independentes e ocorrerem de forma constante ao longo do intervalo de observação.

A função de probabilidade da distribuição de Poisson é dada por:

$$P(X = k) = \frac{{e^{-\lambda} \cdot \lambda^k}}{{k!}} $$

onde:
- $P(X = k) $ é a probabilidade de ocorrerem exatamente $ k $ eventos raros.
- $ e $ é o número de Euler, aproximadamente 2.71828.
- $ \lambda $ é a taxa média de ocorrência dos eventos raros.
- $ k $ é o número de eventos raros que queremos calcular a probabilidade.
- $ k! $ é o fatorial de $ k $.

Essa fórmula nos permite calcular a probabilidade de ocorrer um número específico de eventos raros dentro do intervalo de observação.

A distribuição de Poisson tem algumas características importantes:
- A média $\lambda$ e a variância $\lambda$ são iguais.
- À medida que a taxa média $\lambda$ aumenta, a distribuição de Poisson se aproxima de uma distribuição normal.

A distribuição de Poisson é aplicada em diversos cenários, como modelagem de chegada de clientes em um sistema de atendimento, contagem de erros em um processo de produção, número de acidentes de trânsito em uma região em um determinado período de tempo, entre outros.

Podemos calcular a probabilidade de ocorrerem exatamente $ k $ eventos raros em um intervalo de observação, utilizando a função de probabilidade da distribuição de Poisson. Por exemplo, suponha que a taxa média de ocorrência de acidentes de trânsito em uma região é de 3 acidentes por dia. Qual é a probabilidade de ocorrerem exatamente 4 acidentes em um dia?

````python
# Parâmetro lambda da distribuição de Poisson
lambda_ = 3

# Gerar uma amostra de valores seguindo a distribuição de Poisson
sample = np.random.poisson(lambda_, size=1000)

# Calcular a contagem de ocorrências de cada valor
unique_values, counts = np.unique(sample, return_counts=True)

# Plot do histograma da distribuição de Poisson
plt.bar(unique_values, counts/len(sample), align='center', linewidth=0, width=0.5)
plt.title('Distribuição de Poisson')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.show()
````

Podemos calcular a assimetria e a curtose da distribuição de Poisson.

````python
from scipy.stats import skew, kurtosis

# Cálculo da assimetria
skewness = skew(sample)

# Cálculo da curtose
kurtosis_value = kurtosis(sample)

print("Assimetria:", skewness)
print("Curtose:", kurtosis_value)
````

## Teorema do Limite Central

O Teorema do Limite Central é um resultado fundamental na teoria das probabilidades e estatística. Ele estabelece que, sob certas condições, a soma de um grande número de variáveis aleatórias independentes e identicamente distribuídas (iid) tende a seguir uma distribuição normal, independentemente da distribuição original dessas variáveis.

De forma mais precisa, o Teorema do Limite Central afirma que, conforme o tamanho da amostra aumenta, a distribuição da média amostral se aproxima de uma distribuição normal. Isso ocorre mesmo que a distribuição original das variáveis individuais não seja necessariamente normal.

As principais condições para a aplicação do Teorema do Limite Central são:
1. As variáveis aleatórias devem ser independentes e identicamente distribuídas (iid), o que significa que cada variável tem a mesma distribuição de probabilidade e não está correlacionada com as outras.
2. O tamanho da amostra deve ser grande o suficiente. Em geral, um tamanho de amostra maior que 30 é considerado adequado para que a aproximação normal seja razoável.

O Teorema do Limite Central tem várias implicações práticas e teóricas importantes. Ele permite que os estatísticos usem a distribuição normal como uma aproximação para a distribuição da média amostral, mesmo quando a distribuição original é desconhecida ou não é normal. Isso facilita a realização de inferências estatísticas e a aplicação de métodos estatísticos em uma ampla variedade de situações.

Além disso, o Teorema do Limite Central é a base para a construção de intervalos de confiança e para o teste de hipóteses estatísticas, que são ferramentas fundamentais na inferência estatística.


Vamos simular isso. A seguir, primeiro vamos plotar o histograma de uma distribuição exponential e depois vamos plotar o histograma da média de 1000 amostras de tamanho 1000 dessa distribuição. Vamos ver que a distribuição da média das amostras se aproxima de uma distribuição normal.

````python
# Parâmetro da distribuição exponencial (taxa)
lambd = 2.0

# Tamanho da amostra para a primeira plot
sample_size = 1000

sample = np.random.exponential(1 / lambd, sample_size)

# Plot da distribuição exponencial
plt.figure(figsize=(5, 4))
plt.hist(sample, bins=50, density=True)
plt.title('Distribuição Exponencial')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.show()
````

E agora simular o teorema de limite central para uma distribuição exponencial.

````python
# Parâmetro da distribuição exponencial (taxa)
lambd = 2.0

# Número de amostras a serem geradas
num_samples = 1000

# Tamanho da amostra para calcular a média
sample_size = 1000

# Armazenar as médias amostrais
sample_means = []

# Gerar amostras e calcular as médias
for _ in range(num_samples):
    sample = np.random.exponential(1 / lambd, sample_size)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)


# Plot do histograma das médias amostrais
plt.figure(figsize=(5, 4))
plt.hist(sample_means, bins=30, edgecolor='black')
plt.xlabel('Média Amostral')
plt.ylabel('Frequência')
plt.title('Teorema do Limite Central - Distribuição Exponencial')
plt.show()
````

## Erro Padrão

O erro padrão (standard error) é uma medida da variabilidade das estimativas de uma estatística em relação às diferentes amostras obtidas a partir da mesma população. É frequentemente utilizado para quantificar a precisão da estimativa da média amostral ou de outras estatísticas amostrais.

O cálculo do erro padrão depende do desvio padrão da população $(\sigma)$ e do tamanho da amostra $(n)$. Para a média amostral, o erro padrão é calculado como:

$$SE = \frac{\sigma}{\sqrt{n}} $$

onde:
- $SE$ é o erro padrão
- $\sigma$ é o desvio padrão da população
- $n$ é o tamanho da amostra

O erro padrão indica o quanto as médias amostrais podem variar em relação à média real da população. Quanto menor o erro padrão, mais precisa é a estimativa da média amostral.

É importante ressaltar que, na prática, muitas vezes não conhecemos o desvio padrão da população ($\sigma$), e nesses casos, utilizamos o desvio padrão amostral $(s)$ como uma estimativa do desvio padrão populacional. O cálculo do erro padrão utilizando o desvio padrão amostral é dado por:

$$ SE = \frac{s}{\sqrt{n}}$$

onde $s$ é o desvio padrão amostral.

O erro padrão é especialmente útil quando estamos realizando inferências estatísticas, como a construção de intervalos de confiança ou a realização de testes de hipóteses, pois nos fornece uma medida da incerteza associada à nossa estimativa.

Em resumo, o erro padrão é uma medida da variabilidade das estimativas de uma estatística em relação às diferentes amostras obtidas da mesma população. É calculado com base no desvio padrão da população ou no desvio padrão amostral, juntamente com o tamanho da amostra. Quanto menor o erro padrão, mais precisa é a estimativa da estatística amostral.

## Estimadores

Em estatística, um estimador é uma função utilizada para estimar um parâmetro desconhecido de uma população com base em informações obtidas de uma amostra. Os estimadores são usados para fazer inferências sobre a população com base nos dados disponíveis.

Existem diferentes tipos de estimadores, cada um com suas próprias propriedades e características. Alguns dos estimadores mais comuns incluem:

1. Estimador Pontual: É um estimador que fornece um único valor como estimativa do parâmetro desconhecido. Por exemplo, a média amostral é um estimador pontual da média populacional.

2. Estimador por Intervalo: É um estimador que fornece um intervalo de valores como estimativa do parâmetro desconhecido. Por exemplo, um intervalo de confiança é um estimador por intervalo que fornece um intervalo de valores dentro do qual o parâmetro populacional é provável que esteja contido.

3. Estimador de Máxima Verossimilhança: É um método estatístico utilizado para estimar os parâmetros desconhecidos de uma distribuição estatística. Ele busca encontrar os valores dos parâmetros que maximizam a verossimilhança dos dados observados. Esse tipo de estimador é frequentemente usado quando a distribuição subjacente dos dados é conhecida.

4. Estimador de Momentos: É um estimador que utiliza os momentos estatísticos dos dados para estimar os parâmetros desconhecidos. Os momentos incluem média, variância, assimetria, curtose, entre outros. Esse tipo de estimador utiliza os momentos amostrais para fornecer estimativas dos momentos populacionais.

A escolha do estimador adequado depende do tipo de parâmetro a ser estimado, da distribuição dos dados e das propriedades desejadas para a estimativa. Cada estimador possui suas próprias vantagens e limitações, e é importante considerar cuidadosamente as características dos dados e os objetivos da análise ao escolher um estimador apropriado.


Um estimador mais eficientes é o estimador sem vies (*unbiased*) com a menor variança.

Viés (*bias*) refere-se à tendência de um estimador de superestimar ou subestimar consistentemente o valor real do parâmetro que está sendo estimado. Um estimador é considerado imparcial (*unbiased*) se, em média, produz uma estimativa que é igual ao valor verdadeiro do parâmetro. 

A eficiência, por outro lado, refere-se à precisão de um estimador. Um estimador eficiente é aquele que produz estimativas com baixa variância, o que significa que é menos provável que produza estimativas distantes do valor real. A eficiência é importante porque determina quantos dados são necessários para obter estimativas precisas. Um estimador menos eficiente requer mais dados para atingir o mesmo nível de precisão de um estimador mais eficiente.

Idealmente, um estimador deve ser imparcial e eficiente. No entanto, na prática, muitas vezes há um compromisso (*trade-off*) entre viés e eficiência, e é importante considerar ambos ao selecionar um estimador para um problema específico. Em geral, a imparcialidade é mais importante quando o objetivo é estimar o valor real do parâmetro com precisão, enquanto a eficiência é mais importante quando o objetivo é obter previsões precisas ou tomar decisões com base nas estimativas.

### Estimativa Pontual:
A estimativa pontual é um tipo de estimativa que fornece um único valor como uma estimativa do parâmetro desconhecido da população. Essa estimativa é baseada em informações coletadas a partir de uma amostra. O objetivo é obter um valor que seja considerado o mais próximo possível do valor real do parâmetro.

Existem diferentes métodos para calcular estimativas pontuais, dependendo do parâmetro sendo estimado. Alguns exemplos comuns de estimativas pontuais incluem a média amostral como estimativa da média populacional, a proporção amostral como estimativa da proporção populacional, e a diferença média amostral como estimativa da diferença média populacional, entre outros.

### Intervalo de Confiança:
O intervalo de confiança é uma estimativa estatística que fornece um intervalo de valores dentro do qual é provável que o parâmetro desconhecido esteja contido. Em vez de fornecer um único valor como uma estimativa pontual, o intervalo de confiança oferece uma faixa de valores plausíveis para o parâmetro.

O intervalo de confiança é construído levando em consideração o nível de confiança desejado, que representa a probabilidade de que o intervalo contenha o verdadeiro valor do parâmetro. Por exemplo, um intervalo de confiança de 95% indica que, em 95% das vezes em que repetimos o processo de amostragem e construção do intervalo, o intervalo conterá o valor real do parâmetro.

O cálculo do intervalo de confiança envolve a utilização de estatísticas amostrais e distribuições de probabilidade específicas, como a distribuição t de Student ou a distribuição normal, dependendo do tamanho da amostra e das suposições sobre a população.

O intervalo de confiança é uma medida de incerteza e fornece uma indicação da precisão da estimativa. Quanto maior o nível de confiança desejado, maior será a amplitude do intervalo, refletindo uma maior incerteza.

Em resumo, a estimativa pontual fornece um único valor como estimativa do parâmetro desconhecido, enquanto o intervalo de confiança fornece uma faixa de valores plausíveis para o parâmetro. Ambos os tipos de estimativa são usados em estatística inferencial para obter informações sobre a população com base em dados de amostra.

Por exemplo, quando falamos que o preço médio de um prato de comida no Rio de Janeiro é algo entre 25 e 35 reais, o valor 30 reais é o valor médio e o intervalo $[25, 35]$ é o intervalo de confiança.  

Nunca podemos ter uma confiança de 100% a menos que medimos a população inteira.

A formula para o intervalo de confiança ( do valor médio da população) com a variança conhecida é

$$
CI = [\hat x - z_{\alpha/2}\frac \sigma  {\sqrt n},\; \hat x + z_{\alpha/2}\frac \sigma {\sqrt n}]
$$

## Função de distribuição cumulativa (CDF) e CDF inversa

### CDF

CDF é acumulativo da função de densidade de probabilidade (PDF). Significa qual é a probabilidade dos eventos abaixo de um determinado valor, por exemplo $x$

$$
F(x) = P(X \le x) = \int_{-\infty}^x f(t) dt
$$

onde $X$ é uma variável aleatória. Obviamente, o PDF pode ser calculado pela derivação de seu CDF:

$$
f(x) = \frac{dF(x)}{dx}
$$

Para um caso discreto, o CDF pode ser calculado como a soma das probabilidades dos eventos abaixo de $x$

$$
F(x) = P(X \le x) = \sum^x_{k=-\infty}p(k)
$$

### CDF inverso

CDF: $F(x)$= probabilidade de que $X$ seja menor ou igual a $x$

CDF inverso: $F^{-1}(p)$ = que valor de $x$ produziria um valor de CDF de $p$?

**Exemplo:**

a distribuição dos graus pode ser modelada como uma dist normal. Vamos supor que o máximo seja 200.

Qual é a probabilidade de um aluno atingir 170 ou **menos**? Que o CDF

```python
from scipy.stats import norm
norm.cdf(170, mu, sigma)
```

Digamos que seja 95% (para este exemplo).

Qual é a probabilidade de pontuação **acima** de 170?

```python
from scipy.stats import norm
1 - norm.cdf(170, mu, sigma)
```

o resultado é 5%.

Agora CDF inverso (ICDF). Qual é a pontuação **máxima** dos **últimos** 95% da turma? a resposta é ICDF é 170

```python
from scipy.stats import norm
norm.ppf(0.95, mu, sigma)
```

**termo confuso**: você está “no percentil 95%” significa que está se saindo *melhor* do que os 95% inferiores. Em outra palavra, sua nota é 170 ou superior.
 

**Outro exemplo:**

A altura média de um grupo é de 170 cm, o desvio padrão é de 7 cm.

Seu amigo diz: “minha altura está no percentil 95”. Quão alto ele é?

$$
F^{-1}(0.95,m=170, sd=7) = 181.5 \;cm
$$

Outro amigo tem 160 cm de altura. Em que percentil ele está?

$$
F(160, m=170, sd=7) = 0.08
$$

Ele está no percentil 0,8. Em outras palavras, apenas 8% das pessoas são mais baixas do que ele.

### Coding

````python
from scipy.stats import norm
np.random.seed(0)

mu = 170
sd = 7

# generate sample from dist
x = norm.rvs(loc=mu, scale=sd, size=100)

# maximum likelihood mean
x.mean()
# maximum likelihood variance
x.var()
# or
((x - x.mean())**2).mean()

# maximum likelihood std
x.std()

# unbiased variance
x.var(ddof=1)
# ((x - x.mean())**2)/(len(x)-1)

# at what height are you in the 95th porcentile?
print(norm.ppf(0.95, loc=mu, scale=sd))

# you are 160cm tall, what percentile are you in?
print(norm.cdf(160, loc=mu, scale=sd))

# you are 180cm tall, what is the probability that someone is taller than you?
print(1 - norm.cdf(180, loc=mu, scale=sd))

# there is a function for 1-CDF. It is called survival function
print(norm.sf(180, loc=mu, scale=sd))
````
